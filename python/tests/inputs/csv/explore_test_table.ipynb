```python
import pandas as pd
```

```python
df = pd.read_csv("test.tsv", sep="\t")
df.head()
```

```python
this_activity = "Activity 2"

is_this_activity = df["activity_pref_label"] == this_activity
activity_info = df[is_this_activity]

not_nan = activity_info["preamble"].notna()

preambles = activity_info[not_nan]
preamble = list(preambles["preamble"].unique())

if len(preamble) > 1 or len(preamble) == 0:
    preamble = ""
else:
    preamble = preamble[0]
    
preamble    
```
